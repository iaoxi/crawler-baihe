## 女生的地域分布情况

```python
from pyecharts.charts import Map
from pyecharts import options as opts

city = df["家乡"].value_counts().index.tolist()
data = df["家乡"].value_counts().values.tolist()

zip_data = [(i,j) for i,j in zip(city,data)]
print(zip_data)

china = (
    Map()
    .add('单身妹子区域分布', [(i,j) for i,j in zip(city,data)], 'china')
    .set_global_opts(visualmap_opts=opts.VisualMapOpts(min_=0, max_=300))
)
china.render_notebook()
```

## 自我介绍的词云

```python
import re

all_words = []
for i, v in enumerate(df["自我介绍"]):
    try:
        result = re.findall("'(.*?)'", df["自我介绍"][i])
    except:
        print('error')
    for k in result:
        all_words.extend(result)
word_count = pd.Series(all_words).value_counts()
print(word_count[:20])
```

```python
print(word_count[:20])
```

```python
word_count = word_count.to_frame().reset_index()
word_count.columns = ["关键词","数量"]
word_count[:20]
```

```python
x = []
for i in zip(word_count["关键词"],word_count["数量"]):
    x.append(i)
x[:5]
```

```python
from pyecharts import options as opts
from pyecharts.charts import WordCloud

wordcloud = (
    WordCloud(init_opts=opts.InitOpts(width='750px', height='350px'))
    .add("", x[:100], word_size_range=[20, 40])
    .set_global_opts(title_opts=opts.TitleOpts(title="自我介绍"))
)
wordcloud.render_notebook()
```

```python
from pyecharts.charts import Page

page = Page(layout=Page.DraggablePageLayout)
page.add(
    bar1,
    bar3,
    china,
    bar4,
    bar2,
    line_man1,
    pie,
    wordcloud
)
page.render_notebook()
```
