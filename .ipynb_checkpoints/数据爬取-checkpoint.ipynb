```python
import requests
import re
import json
import time
from lxml import etree
import pandas as pd
import warnings
warnings.filterwarnings("ignore")

# 请求头，伪装成浏览器
#
HEADERS = {
    # 添加你自己的Cookie
    'Cookie': '',
    'Origin': 'https://search.baihe.com',
    'Pragma': 'no-cache',
    'Referer': 'https://search.baihe.com/',
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.116 Safari/537.36',
}
# 这后来发现没用了
temp_url = 'https://search.baihe.com/Search/getUserID?&jsonCallBack?'
# 主页
tql_url = 'https://search.baihe.com/'
# 个人信息的前url
info_url = 'https://profile1.baihe.com/?oppID='
# 获得id集合url
url_base = 'https://search.baihe.com/Search/getUserID?&jsonCallBack=jQuery1830923921797491073_1594465799055'
session = requests.Session()
# 不用持续连接
session.keep_alive = False
# 从主页中获得一些Cookie等信息
session.get(url=tql_url, headers=HEADERS)




# 爬取的页数
for page in range(2, 66):
    print(f'正在爬取第{page}页数据......')
    data = {
        'page': page
    }
    url_tails = session.post(url=url_base, headers=HEADERS, data=data).content.decode('utf-8')
    # 提取规范的JSON格式的字符串
    json_data = re.search(r'jQuery1830923921797491073_1594465799055\((.*)\);', url_tails, flags=0)
    json_data_format = json.loads(json_data.group(1))
    data_uid = json_data_format['data']
    # 需要解析的数据
    me_uid = []
    me_age = []
    me_height = []
    me_education = []
    me_salary = []
    me_location = []
    me_marriage = []
    me_home = []
    me_introduce = []
    he_age = []
    he_height = []
    he_education = []
    he_salary = []
    he_location = []
    he_marriage = []
    he_home = []
    # 遍历uid来获得个人信息页
    for uid in data_uid:
        try:
            # 拼接个人信息页完整url
            me_url = info_url + uid
            me_page = session.get(me_url, headers=HEADERS)
            me_page.decoding = 'gbk'
            me_page = me_page.text
            # 这边有些异常编码的参数需要处理
            me_page = me_page.encode('GBK','ignore').decode('GBk')
            tree = etree.HTML(me_page)
            # me_uid.append(uid)
            
            me_age.append(tree.xpath('//*[@id="matching_detail"]/div/div/dl[2]/dd[1]/text()')[0])
            me_height.append(tree.xpath('//*[@id="matching_detail"]/div/div/dl[3]/dd[1]/text()')[0])
            me_education.append(tree.xpath('//*[@id="matching_detail"]/div/div/dl[4]/dd[1]/text()')[0])
            me_salary.append(tree.xpath('//*[@id="matching_detail"]/div/div/dl[5]/dd[1]/text()')[0])
            me_location.append(tree.xpath('//*[@id="matching_detail"]/div/div/dl[8]/dd[1]/text()')[0])
            me_marriage.append(tree.xpath('//*[@id="matching_detail"]/div/div/dl[6]/dd[1]/text()')[0])
            me_home.append(tree.xpath('//*[@id="matching_detail"]/div/div/dl[7]/dd[1]/text()')[0])
            me_introduce.append(tree.xpath('//*[@id="profileCommon"]/div[1]/div[2]/div[1]/text()')[0])
            he_age.append(tree.xpath('//*[@id="matching_detail"]/div/div/dl[2]/dd[3]/text()')[0])
            he_height.append(tree.xpath('//*[@id="matching_detail"]/div/div/dl[3]/dd[3]/text()')[0])
            he_education.append(tree.xpath('//*[@id="matching_detail"]/div/div/dl[4]/dd[3]/text()')[0])
            he_salary.append(tree.xpath('//*[@id="matching_detail"]/div/div/dl[5]/dd[3]/text()')[0])
            he_location.append(tree.xpath('//*[@id="matching_detail"]/div/div/dl[8]/dd[3]/text()')[0])
            he_marriage.append(tree.xpath('//*[@id="matching_detail"]/div/div/dl[6]/dd[3]/text()')[0])
            he_home.append(tree.xpath('//*[@id="matching_detail"]/div/div/dl[7]/dd[3]/text()')[0])
        except IndexError:
            print(IndexError)
        print(f'{uid}，已经爬取成功！')
        time.sleep(0.5)
    df = pd.DataFrame()
    df["年龄"] = me_age
    df["身高"] = me_height
    df["学历"] = me_education
    df["薪资"] = me_salary
    df["家乡"] = me_location
    df["婚姻"] = me_marriage
    df["住房情况"] = me_home
    df["自我介绍"] = me_introduce
    df["对象年龄"] = he_age
    df["对象身高"] = he_height
    df["对象学历"] = he_education
    df["对象薪资"] = he_salary
    df["对象家乡"] = he_location
    df["对象婚姻"] = he_marriage
    df["对象住房情况"] = he_home
    # 保存文件
    try:
        df.to_csv("baihe.csv", mode="a+", header=None, index=None, encoding="gbk")
    except:
        print('保存数据异常')
    # 慢点 防止 封IP
    time.sleep(1)
    print(f'第{page}页数据爬取完毕，即将爬取下一页......')
    time.sleep(10)

print("数据爬取完毕，离脱单不远啦！")
```
